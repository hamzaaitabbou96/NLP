# Removing stop words with SpaCy 

##### The SpaCy stop word list is basically a set of strings. You can add a new word to the set like you would add any new item to a set.

```
import spacy
# nltk for word_tokenize
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize 

# load the core English language model
nlp = spacy.load('en_core_web_sm')
```

```
all_stopwords = nlp.Defaults.stop_words

# input text
text = "hamza likes to play football, however he is not too fond of tennis."
text_tokens = word_tokenize(text)
tokens_without_sw= [word for word in text_tokens if not word in all_stopwords]

print("with stop words : \n",text_tokens) 
print('-_-')
print("without stop words : \n",tokens_without_sw) 
```

### Adding and Removing Stop Words in SpaCy Default Stop Word List

```
my_stopwords = {'This','play','however','many','Many','so','cant','These','these'}
all_stopwords |= my_stopwords
```

```
# input text
text = "hamza likes to play football, however he is not too fond of tennis."
text_tokens = word_tokenize(text)
tokens_without_sw= [word for word in text_tokens if not word in all_stopwords]

print("with stop words : \n",text_tokens) 
print('-_-')
print("without stop words : \n",tokens_without_sw)
```
